1. What is a random variable in probability theory?
- A random variable is a numerical outcome of a random process or experiment. It assigns a real number to each possible outcome in a sample space.

2. What are the types of random variables?
- Different type of random variable are :
 - Discrete random variable: Takes countable values (e.g., 0, 1, 2, ‚Ä¶).

 - Continuous random variable: Takes any value in a continuous range (e.g., height, temperature).



3. What is the difference between discrete and continuous distributions?
- The difference between discrete and continuous distributions are :
 - Discrete distributions deal with discrete random variables and assign probabilities to specific values.

 - Continuous distributions deal with continuous variables and use a probability density function; probabilities are determined over intervals.

4. What are probability distribution functions (PDF)?
- A Probability Distribution Function (PDF) describes the likelihood of a random variable taking specific values. For discrete variables, it's a probability mass function (PMF), and for continuous variables, it‚Äôs a probability density function (pdf).

5. How do cumulative distribution functions (CDF) differ from PDFs?
- PDF shows the probability of a specific value (or range for continuous).

- CDF shows the cumulative probability up to a certain value:

6. What is a discrete uniform distribution?

9. What is the Poisson distribution and where is it applied?
- The Poisson distribution models the number of events in a fixed interval of time or space, given a constant average rate Œª.

 Used in: call centers, traffic flow, etc.



10. What is a continuous uniform distribution?
- A continuous uniform distribution is a probability distribution where a continuous random variable has an equal probability of taking any value within a specified range (between a lower limit 'a' and an upper limit 'b'). Essentially, all outcomes within that interval are equally likely, and the probability density function is constant over that range.

11. What are the characteristics of a normal distribution?
- Characteristics of a normal distribution are :
 - Bell-shaped and symmetric

 - Mean = Median = Mode

 - Defined by mean (Œº) and standard deviation (œÉ)

12. What is the standard normal distribution, and why is it important?
- It‚Äôs a normal distribution with mean = 0 and standard deviation = 1.
Important because:

 - Used for calculating Z-scores

 - Simplifies statistical testing and comparisons



13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The CLT states that the distribution of sample means approximates a normal distribution as sample size increases, regardless of the population's distribution.
Critical for:

 - Justifying use of normal approximation

 - Enabling hypothesis testing



14. How does the Central Limit Theorem relate to the normal distribution?
- It allows us to use the normal distribution to estimate the behavior of sample means, even when the underlying population is not normally distributed (as long as n is large).



15. What is the application of Z statistics in hypothesis testing?
- Z-statistics are used to test hypotheses when the population variance is known and/or the sample size is large. It helps determine how far the sample mean is from the population mean in terms of standard deviations.



16. How do you calculate a Z-score, and what does it represent?
- Z-score represents how many standard deviations a data point is from the mean.
  It is calculated by formula :

    Z = X-Œº/œÉ

 Where:
  - X = sample value

17. What are point estimates and interval estimates in statistics?
- Point estimate: A single value estimate of a parameter (e.g., sample mean).

- Interval estimate: A range within which the parameter lies, usually with a confidence level (e.g., confidence interval).



18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals (CI) show the range of plausible values for a population parameter and indicate the reliability of the estimate. A 95% CI means there's a 95% chance the true value lies within that interval.



19. What is the relationship between a Z-score and a confidence interval?
- Relationship between a Z-score and a confidence interval are :
 - A Z-score measures how many standard deviations a data point is from the mean.
 - A confidence interval provides a range of values likely to contain a population parameter.

20. How are Z-scores used to compare different distributions?
- Z-scores standardize values across different distributions by converting them to the same scale (mean = 0, SD = 1), making comparisons meaningful.



21. What are the assumptions for applying the Central Limit Theorem?
- Assumptions to apply on the Central Limit Theorem are :
 - Independent random samples

 - Identically distributed variables

 - Sufficiently large sample size (typically n ‚â• 30)

 - Finite mean and variance



22. What is the concept of expected value in a probability distribution?
- The expected value (also known as mean or mathematical expectation) of a random variable is the long-run average value it takes over many repetitions of an experiment or trial. It gives a measure of the center of a probability distribution.



23. How does a probability distribution relate to the expected outcome of a random variable?
- A probability distribution provides the probabilities for each outcome, and the expected value summarizes the average or ‚Äúexpected‚Äù outcome based on that distribution.

```python
#2. What are the types of random variables.
import matplotlib.pyplot as plt
from scipy.stats import randint

x = np.arange(1, 7)
pmf = randint.pmf(x, 1, 7)
plt.stem(x, pmf)
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("x")
plt.ylabel("PMF")
plt.grid()
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    probs = bernoulli.pmf(x, p)
    return dict(zip(x, probs))

print("Bernoulli PDF (p=0.6):", bernoulli_pdf(0.6))

```

```python
# 4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
from scipy.stats import binom
import matplotlib.pyplot as plt


data = binom.rvs(n=10, p=0.5, size=1000)
plt.hist(data, bins=11, edgecolor='black', density=True)
plt.title("Binomial Distribution Histogram")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()

```

```python
# 5. Create a Poisson distribution and visualize it using Python.
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson

x = np.arange(0, 20)
pmf = poisson.pmf(x, mu=5)
plt.stem(x, pmf)
plt.title("Poisson Distribution (Œª=5)")
plt.xlabel("x")
plt.ylabel("PMF")
plt.grid()
plt.show()
```

```python
# 6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.cdf = randint.cdf(x, 1, 7)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint
x = np.arange(1, 7)
cdf = randint.cdf(x, 1, 7)


plt.step(x, cdf)
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("x")
plt.ylabel("CDF")
plt.grid()
plt.show()


```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.
sample_mean = 72
pop_mean = 70
std_dev = 10
n = 30

z = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
print(f"Z-statistic: {z}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.
import scipy.stats as stats

data = np.random.normal(50, 5, 100)
mean = np.mean(data)
sem = stats.sem(data)
conf = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)
print("95% Confidence Interval:", conf)

```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
data = np.random.normal(60, 8, 100)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)
print("95% Confidence Interval:", ci)

```

```python
# 21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.
data1 = np.random.normal(10, 2, 100)
data2 = np.random.normal(10, 5, 100)

std1 = np.std(data1)
std2 = np.std(data2)

plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'])
plt.title("Comparison of Standard Deviations")
plt.show()

print("Std Dev 1:", std1, "Std Dev 2:", std2)

```

```python
# 22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
data = np.random.normal(50, 10, 100)
data_range = np.max(data) - np.min(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
print("Range:", data_range, "IQR:", iqr)

```
